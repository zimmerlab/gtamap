```python
# print all occurrences of the sequence "TCG"
def all_occurrences(seq, sub):
    start = 0
    while True:
        start = seq.find(sub, start)
        if start == -1:
            break
        yield start
        start += len(sub)

# print all occurrences of "TCG" in tas2r4_seq_fw
for i in all_occurrences(tas2r4_seq_fw, "CTTCC"):
    print(i)
```
